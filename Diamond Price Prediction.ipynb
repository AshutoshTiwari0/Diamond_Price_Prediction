```python
import matplotlib.pyplot as plt
import numpy as np

# Example data - replace these with your actual values
models = ['Linear', 'ElasticNet', 'Ridge', 'Lasso']
r2_scores = [0.87, 0.90, 0.99, 1.00]  # Replace with your RÂ² scores
rmse_scores = [1438.94, 1137.92, 12.38,0.0]    # Replace with your RMSE scores

# Creating the bar plot for RÂ² scores
fig, ax1 = plt.subplots(figsize=(10, 6))

bar_width = 0.35
index = np.arange(len(models))

bar1 = ax1.bar(index, r2_scores, bar_width, label='RÂ² Score', color='b')

# Creating the bar plot for RMSE scores
ax2 = ax1.twinx()
bar2 = ax2.bar(index + bar_width, rmse_scores, bar_width, label='RMSE', color='g')

# Adding labels, title, and custom x-axis tick labels
ax1.set_xlabel('Model')
ax1.set_ylabel('RÂ² Score', color='b')
ax2.set_ylabel('RMSE', color='g')
ax1.set_title('Comparison of Linear Regression Models')
ax1.set_xticks(index + bar_width / 2)
ax1.set_xticklabels(models)

# Adding legend
bars = bar1 + bar2
labs = [b.get_label() for b in bars]
ax1.legend(bars, labs, loc='upper right')

# Displaying the plot
plt.show()
```
